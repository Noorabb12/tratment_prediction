```python
#Encoding the data
enc=LabelEncoder()
for x in data.columns:
    if data[x].dtype=='object':
        data[x]=enc.fit_transform(data[x])
data.info()
```

```python
y=data['SOURCE']
X=data.drop(['SOURCE'],axis=1)
```

### PCA 


```python
# Initialize PCA with the number of components you want to retain
n_components = 2
pca = PCA(n_components=n_components)

# Fit PCA to your data
principal_components = pca.fit_transform(X)

# Create a DataFrame to store the principal components
principal_df = pd.DataFrame(data=principal_components, columns=[f'PC{i+1}' for i in range(n_components)])

# Print the explained variance ratio to understand the amount of variance retained by each component
print("Explained Variance Ratios:")
print(pca.explained_variance_ratio_)

# You can use principal_df for further analysis or modeling with reduced dimensions
print("Data after PCA:")
print(principal_df.head())
```

```python
# Convert to dataframe
component_names = [f"PC{i+1}" for i in range(principal_components.shape[1])]
```

```python
loadings = pd.DataFrame(
    pca.components_.T,  # transpose the matrix of loadings
    columns=component_names,  # so the columns are the principal components
    index=X.columns,  # and the rows are the original features
)
loadings
```

```python
from sklearn.feature_selection import mutual_info_regression
def plot_variance(pca, width=8, dpi=100):
    # Create figure
    fig, axs = plt.subplots(1, 2)
    n = pca.n_components_
    grid = np.arange(1, n + 1)
    # Explained variance
    evr = pca.explained_variance_ratio_
    axs[0].bar(grid, evr)
    axs[0].set(
        xlabel="Component", title="% Explained Variance", ylim=(0.0, 1.0)
    )
    # Cumulative Variance
    cv = np.cumsum(evr)
    axs[1].plot(np.r_[0, grid], np.r_[0, cv], "o-")
    axs[1].set(
        xlabel="Component", title="% Cumulative Variance", ylim=(0.0, 1.0)
    )
    # Set up figure
    fig.set(figwidth=8, dpi=100)
    return axs

def make_mi_scores(X, y, discrete_features):
    mi_scores = mutual_info_regression(X, y, discrete_features=discrete_features)
    X = pd.DataFrame(X)
    mi_scores = pd.Series(mi_scores, name="MI Scores", index=X.columns)
    mi_scores = mi_scores.sort_values(ascending=False)
    return mi_scores
```

```python
# Look at explained variance
plot_variance(pca);
```

```python
mi_scores = make_mi_scores(principal_components, y, discrete_features=False)
mi_scores
```

```python
# plotting confusion matrix
import itertools
def plot_confusion_matrix(cm, classes, title='Confusion Matrix', cmap=plt.cm.Blues):
    """
    This function plots a confusion matrix.
    
    Parameters:
        cm (array-like): Confusion matrix as returned by sklearn.metrics.confusion_matrix.
        classes (list): List of class names, e.g., ['Class 0', 'Class 1'].
        title (str): Title for the plot.
        cmap (matplotlib colormap): Colormap for the plot.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    
    fmt = '.0f'
    thresh = cm.max() / 2.0
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt), horizontalalignment="center", color="white" if cm[i, j] > thresh else "black")
    
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()


# Plot the confusion matrix
class_names = ['out', 'in'] # positive class should come the last
plot_confusion_matrix(confusion, class_names)
```
